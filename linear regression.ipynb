```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from numpy.polynomial.polynomial import polyfit
from sklearn import linear_model
```

```python
#2.5 * np.random.randn(2,4)+3  #2.5是标准差，3是期望
df = pd.DataFrame(2.5 * np.random.randn(10,2)+3 ,
                   columns=['x', 'y'])
x=df.iloc[:,0:1]
y=df.iloc[:,1:2]
```

```python
reg = linear_model.LinearRegression()
fit = reg.fit(x,y)
fit.coef_
fit.intercept_
plt.plot(x,y,'o', label='data')
plt.plot(x,fit.intercept_+fit.coef_*x,label='regression')
plt.legend()
plt.show()
```
